```python
df = pd.read_csv('/kaggle/input/clinic/ultima_all_clean.csv')
df = df.set_index('hn')
df_clean_column = df.drop(['pchemo','surgery','size','appearance','Wardsize','RHlvsi','depth','nodeyiel','vgmargin','pelvicme','adnmet','utmet','vgmet'],axis = 1)
df_clean = df_clean_column.dropna(axis = 0)
```

```python
y = df_clean['pmmet'].values
X = df_clean.drop(['pmmet'],axis = 1).values

from imblearn.over_sampling import SMOTE
sm = SMOTE(random_state=1)
X, y = sm.fit_resample(X, y)

from sklearn.model_selection import train_test_split
X, Xt, y, yt = train_test_split(X, y, test_size=0.2, random_state=0)

from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2
fs = SelectKBest(chi2, k=5)
fs.fit(X, y)
X = fs.transform(X)
Xt = fs.transform(Xt)

from sklearn.decomposition import PCA
fr = PCA(n_components=3)
fr.fit(X)
X = fr.transform(X)
Xt = fr.transform(Xt)
```
